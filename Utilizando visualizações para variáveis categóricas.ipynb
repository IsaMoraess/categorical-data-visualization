```python
# bibliotecas para analise
import pandas as pd
# biblioteca para visualizar graficos
import seaborn as sns
# comunica com o seaborn
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
# iris tem tres sub-especies
iris['species'].value_counts()
```

```python
iris.describe()
```

```python
# gerar um grafico (histograma)
col = 'sepal_length'
sns.histplot(data = iris, x = 'sepal_length').set_title(f'Distribuição de variavel {col}')
```

```python
# mostra a curva da distribuição (kde)
col = 'sepal_length'
sns.histplot(data = iris, x = 'sepal_length', kde = True).set_title(f'Distribuição de variavel {col}')
```

```python
# ver mais de um garfico de uma vez so (plt.show())
for col in iris.drop(columns = 'species'):
    sns.histplot(data = iris, x = col, kde = True).set_title(f'Distribuição de variavel {col}')
    plt.show()
```

```python
# passando a coloração cm iris especies (grafico cm separação por flor)
for col in iris.drop(columns = 'species'):
    sns.histplot(data = iris, x = col, kde = True, hue = iris['species']).set_title(f'Distribuição de variavel {col}')
    plt.show()
```

```python
# media do tamanho, ci = intervalo d econfiança
for col in iris.drop(columns = 'species'):
    sns.barplot(data = iris, x = iris['species'], hue = iris['species'], y = col, ci = 90)
    plt.show()
```
